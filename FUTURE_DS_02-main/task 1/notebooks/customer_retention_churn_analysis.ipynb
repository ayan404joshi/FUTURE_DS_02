# Customer Retention & Churn Analysis

This notebook is a lightweight companion to the production pipeline. It reads the generated analysis artifacts, summarizes the key metrics, and points to the saved charts and report outputs.

```python
from pathlib import Path
import json

root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
summary_path = root / 'data' / 'processed' / 'analysis_summary.json'
model_path = root / 'data' / 'processed' / 'model_metrics.json'

summary = json.loads(summary_path.read_text())
model = json.loads(model_path.read_text())

print(f"Customers analyzed: {summary['total_customers']:,}")
print(f"Overall churn rate: {summary['churn_rate']:.1%}")
print(f"Average CLV: ${summary['avg_clv']:.2f}")
print(f"Best model: {model['best_model']}")
print(f"Best ROC-AUC: {model['metrics'][model['best_model']]['roc_auc']:.3f}")
```
